```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk menampilkan histogram
def plot_histogram(image, title):
    hist = cv2.calcHist([image], [0], None, [256], [0,256])
    plt.plot(hist, label=title)
    plt.xlim([0,256])
    plt.legend()

```

```python
# Load citra grayscale (misalnya X-ray atau MRI)
image = cv2.imread('WhatsApp Image 2025-11-12 at 11.16.14.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Normalisasi kontras dengan histogram equalization
equalized = cv2.equalizeHist(image)

```

```python
# Tampilkan citra asli dan hasil normalisasi
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(equalized, cmap='gray')
plt.title('Citra Setelah Normalisasi')
plt.axis('off')
plt.show()
```
